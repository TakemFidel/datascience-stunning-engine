```python
# Load the input text file
lines = sc.textFile("welcome.txt")
```

```python
# Split each line into words
words = lines.flatMap(lambda line: line.split(" "))
```

```python
# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))
```

```python
# Reduce by key to get the count of each word
word_counts = word_counts.reduceByKey(lambda a, b: a + b)
```

```python
# Collect the results and print them
results = word_counts.collect()
for result in results:
    print(result)
```

## PySpark Analysis & Statistics

```python
from pyspark.sql import SparkSession
import pandas as pd
import matplotlib.pyplot as plt

spark = SparkSession.builder.appName("ShopCam_Analytics").getOrCreate()

# Load to Spark
df_c = spark.createDataFrame(pd.DataFrame(list(db.clients.find())))
df_o = spark.createDataFrame(pd.DataFrame(list(db.commandes.find())))
df_p = spark.createDataFrame(pd.DataFrame(list(db.produits.find())))

# Task: Total revenue (turnover) by city
# Join orders with products to get price, then with clients to get city
df_revenue = df_o.join(df_p, df_o.produit_id == df_p._id) \
                 .withColumn("total", df_o.quantite * df_p.prix) \
                 .join(df_c, df_o.client_id == df_c._id) \
                 .groupBy("ville").sum("total")

df_revenue.show()

# Export results to CSV
df_revenue.toPandas().to_csv("turnover_by_city.csv", index=False)
print("💾 Results exported to turnover_by_city.csv")
```
